## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
data
```

```python
json_normalize(data,'counties',['state','shortname',['info','governor']])
```

```python
df = json_normalize(data,'counties',['state','shortname',['info','governor']])
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
jgt = json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
jgt.sort_values('population').tail(2)
```

```python
df1 = jgt[[ 'population', 'name']]
df2 = df1.sort_values('population')
df2.head(5)
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json_df = pd.read_json('data/world_bank_projects.json')
json_df.head()
```

```python
#Question 1, Find the 10 countries with most projects
json_df.countryname.value_counts().head(10)
```

```python
#Answer to Question 1
#Africa is not a country so I dropped it.
json_df.countryname.value_counts().drop('Africa').head(10)
```

```python
#Answer to Question 2
#Find the top 10 major project themes (using column 'mjtheme_namecode')
json_df2 = json.load(open('data/world_bank_projects.json'))
json_normalize(json_df2, 'mjtheme_namecode', 'project_name').head(10)
```

```python
#Question 3, In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.
jf3 = json_normalize(json_df2, 'mjtheme_namecode', 'project_name')
df3 = jf3.sort_values('name')
df3
```

```python
#created a dataframe with only the missing values
df4 = df3[df3['name'] == df3.iloc[0]['name']]
df4
```

```python
#Answer to Question #3
#create and fill in the missing values of the original data frame 
df3.replace([df3.iloc[0]['name']], [np.nan])
```
